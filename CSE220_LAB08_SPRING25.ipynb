```python
#Task 03



def distribute_tasks(tasks, m):

    loads = [0] * m


    for task in tasks:

        min_index = 0
        for i in range(1, m):
            if loads[i] < loads[min_index]:
                min_index = i

        loads[min_index] += task


    return loads


tasks = [2, 4, 7, 1, 6]
m = 4


result = distribute_tasks(tasks, m)
print(result)
```

```python
#Task 04

import numpy as np

class MaxHeap:
    def __init__(self, max_size):
        self.heap = np.full(max_size + 1, -999, dtype=int)
        self.size = 0

    def swim(self, idx):
        if idx == 1:
            return
        parent = idx // 2
        if self.heap[parent] < self.heap[idx]:
            self.heap[parent], self.heap[idx] = self.heap[idx], self.heap[parent]
            self.swim(parent)

    def insert(self, key):
        self.size += 1
        self.heap[self.size] = key
        self.swim(self.size)

    def sink(self, i):
        le = 2 * i
        ra = 2 * i + 1
        idx = i

        if le <= self.size and self.heap[le] > self.heap[idx]:
            idx = le
        if ra <= self.size and self.heap[ra] > self.heap[idx]:
            idx = ra

        if idx != i:
            self.heap[i], self.heap[idx] = self.heap[idx], self.heap[i]
            self.sink(idx)

    def extractMax(self):
        if self.size == 0:
            return None
        max_val = self.heap[1]
        self.heap[1] = self.heap[self.size]
        self.heap[self.size] = -999
        self.size -= 1
        self.sink(1)
        return max_val

    def getHeap(self):
        return self.heap[1:self.size + 1]


def top_k_largest(nums, k):

    heap = MaxHeap(len(nums))


    for num in nums:
        heap.insert(num)


    result = []
    for _ in range(k):
        result.append(heap.extractMax())

    return result

nums = [4, 10, 2, 8, 6, 7]
k = 3


result = top_k_largest(nums, k)
print("Top k Largest Elements:", result)
```
